```
data_list = [(100, "Prashant", 45, 45000),
             (101, "Tarun", 36, 33000),
             (102, "David", 48, 28000)]

test_df = spark.createDataFrame(data_list).toDF("id", "name", "age", "salary")
display(test_df)
```

```
from pyspark.sql.functions import expr
test_df1 = test_df.withColumn("increment", expr("salary * 10/100"))
display(test_df1)
```

```
test_df2 = test_df1.withColumn("salary", expr("salary + increment"))
display(test_df2)
```

```
data_list = [(101, "Mumbai", "Goa", 587),
             (102, "Mumbai", "Bangalore", 985),
             (102, "Mumbai", "Bangalore", 985),
             (103, "Delhi", "Chennai", 2208),
             (104, "Delhi", "Chennai", 2208),
             (105, "Bangalore", "Kolkata", 1868),
             (105, "Bangalore", "Kolkata", 1865)]

df = spark.createDataFrame(data_list) \
    .toDF("id", "source", "destination", "distance")

display(df)
```

```
df.filter("source like 'Mum%'") \
  .where("destination like 'Ban%'") \
  .show()
```

```
from pyspark.sql.functions import col
df.filter(col("source").like("Mum%")) \
  .where(col("destination").like("Ban%")) \
  .show()
```

```
df.filter("source like 'Mum%' and destination like 'Ban%'") \
  .show()
```

```
from pyspark.sql.functions import col
df.filter(col("source").like("Mum%") & col("destination").like("Ban%")) \
  .show()
```

```
df.filter("source like 'Mum%' or destination like 'Ban%'") \
  .show()
```

```
from pyspark.sql.functions import col
df.filter(col("source").like("Mum%") | col("destination").like("Ban%")) \
  .show()
```

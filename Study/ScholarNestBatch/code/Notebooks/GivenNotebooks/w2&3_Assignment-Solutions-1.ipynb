##### 1. Create a Spark data frame using the /databricks-datasets/learning-spark-v2/sf-fire/sf-fire-calls.csv

```
raw_fire_df = spark.read \
            .format("csv") \
            .option("header", "true") \
            .option("inferSchema", "true") \
            .load("/databricks-datasets/learning-spark-v2/sf-fire/sf-fire-calls.csv")
```

##### 2. Create a Global Temporary View using the above data frame. Assume the view name is fire_service_calls_view.

```
raw_fire_df.createGlobalTempView("fire_service_calls_view")
```

##### 3. Verify your view is working

```
%sql
select * from global_temp.fire_service_calls_view
```

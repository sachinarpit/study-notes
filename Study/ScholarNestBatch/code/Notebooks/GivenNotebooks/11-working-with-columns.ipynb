```
airlines_df = spark.read \
                .format("csv") \
                .option("header", "true") \
                .option("inferSchema","true") \
                .option("samplingRatio", "0.0001") \
                .load("/databricks-datasets/airlines/part-00000")

display(airlines_df)
```

```
airlines_df.select("Origin", "Dest", "Distance", "ArrDelay", "DepDelay" ) \
           .show(5)
```

```
from pyspark.sql.functions import expr, column, col

airlines_df.select("Origin", expr("Dest"), airlines_df.Distance, column("ArrDelay"), col("DepDelay") ) \
           .show(5)
```

```
airlines_df.select("Origin", "Dest", expr("Distance * 1.6 as km_distance") ) \
           .show(5)
```

```
airlines_df.select("UniqueCarrier", 
                   "FlightNum", 
                   expr("Year||'-'||Month||'-'||DayofMonth as FlightDate"), 
                   expr("Distance * 1.6 as DistanceKM")) \
            .show(5)
```

```
data_list = [("SAN", "San Antonio"),
             ("SFO", "San Francisco"),
             ("BUR", "Hollywood Burbank"),
             ("OAK", "Oakland")]
airport_df = spark.createDataFrame(data_list).toDF("Origin", "OriginAirportName")
display(airport_df)
```

```
airlines_df.join(airport_df, airlines_df.Origin==airport_df.Origin, "left") \
           .select(airlines_df.Origin, "OriginAirportName", "Dest", "Distance") \
           .where("Origin in ('SAN', 'SFO', 'BUR', 'OAK')") \
           .distinct() \
           .show(5)
```

```
cFactor = 1.6
airlines_df.select("Origin", "Dest", expr("Distance * {} as km_distance".format(cFactor)) ) \
           .show(5)
```

```
cFactor = 1.6
airlines_df.select("Origin", "Dest", col("Distance") * cFactor)  \
           .show(5)
```

```
airlines_df.select("Origin", "Dest", "Distance","ArrDelay", "DepDelay") \
           .where((col("Origin").like("B%")) & 
                  (col("ArrDelay") < 0)) \
           .show(5)
```

```
airlines_df.select("Origin", "Dest", "Distance","ArrDelay", "DepDelay") \
           .where("Origin like 'B%' and ArrDelay < 0") \
           .show(5)
```
